```python
# ==================================
# STEP 2: FIX DATA TYPES
# ==================================

import pandas as pd

# -------- BOOKINGS TABLE --------
if 'booking_date' in Bookings_08_df.columns:
    Bookings_08_df['booking_date'] = pd.to_datetime(Bookings_08_df['booking_date'], errors='coerce')

if 'travel_date' in Bookings_08_df.columns:
    Bookings_08_df['travel_date'] = pd.to_datetime(Bookings_08_df['travel_date'], errors='coerce')

if 'total_amount' in Bookings_08_df.columns:
    Bookings_08_df['total_amount'] = pd.to_numeric(Bookings_08_df['total_amount'], errors='coerce')

if 'num_people' in Bookings_08_df.columns:
    Bookings_08_df['num_people'] = pd.to_numeric(Bookings_08_df['num_people'], errors='coerce')


# -------- PAYMENTS TABLE --------
if 'payment_date' in Payments_09_df.columns:
    Payments_09_df['payment_date'] = pd.to_datetime(Payments_09_df['payment_date'], errors='coerce')

if 'amount_paid' in Payments_09_df.columns:
    Payments_09_df['amount_paid'] = pd.to_numeric(Payments_09_df['amount_paid'], errors='coerce')


# -------- REVIEWS TABLE --------
if 'rating' in Reviews_10_df.columns:
    Reviews_10_df['rating'] = pd.to_numeric(Reviews_10_df['rating'], errors='coerce')


# -------- PACKAGES TABLE --------
if 'price' in Packages_11_df.columns:
    Packages_11_df['price'] = pd.to_numeric(Packages_11_df['price'], errors='coerce')

if 'duration_days' in Packages_11_df.columns:
    Packages_11_df['duration_days'] = pd.to_numeric(Packages_11_df['duration_days'], errors='coerce')


# -------- FLIGHTS TABLE --------
if 'departure_time' in Flights_Table_05_df.columns:
    Flights_Table_05_df['departure_time'] = pd.to_datetime(Flights_Table_05_df['departure_time'], errors='coerce')

if 'arrival_time' in Flights_Table_05_df.columns:
    Flights_Table_05_df['arrival_time'] = pd.to_datetime(Flights_Table_05_df['arrival_time'], errors='coerce')


# -------- HOTEL ROOMS --------
if 'room_price' in HotelRooms_13_df.columns:
    HotelRooms_13_df['room_price'] = pd.to_numeric(HotelRooms_13_df['room_price'], errors='coerce')


# -------- CAR RENTALS --------
if 'rental_price' in CarRentals_18_df.columns:
    CarRentals_18_df['rental_price'] = pd.to_numeric(CarRentals_18_df['rental_price'], errors='coerce')


print("✅ Data Type Fixing Completed Successfully")
```

```python
# ==========================================
# STEP 3: MERGING DATA
# ==========================================

# 1️⃣ Standardize all column names to lowercase
for df in [Bookings_08_df, Customers_07_df, Payments_09_df,
           Packages_11_df, Destinations_Table_03_df,
           Cities_Table_02_df, Countries_Table_01_df]:
    df.columns = df.columns.str.lower()


# 2️⃣ Function to safely merge without duplicate column conflict
def safe_merge(left_df, right_df, key):

    # Remove duplicate columns from right except the key
    common_cols = set(left_df.columns).intersection(set(right_df.columns))
    common_cols = common_cols - {key}

    right_df_clean = right_df.drop(columns=list(common_cols), errors='ignore')

    return left_df.merge(right_df_clean, on=key, how="left")


# Start merging step by step

# Bookings + Customers
master_df = safe_merge(Bookings_08_df, Customers_07_df, "customer_id")

# Payments
master_df = safe_merge(master_df, Payments_09_df, "booking_id")

# Packages
if "package_id" in master_df.columns and "package_id" in Packages_11_df.columns:
    master_df = safe_merge(master_df, Packages_11_df, "package_id")

# Destinations
if "destination_id" in master_df.columns and "destination_id" in Destinations_Table_03_df.columns:
    master_df = safe_merge(master_df, Destinations_Table_03_df, "destination_id")

# Cities
if "city_id" in master_df.columns and "city_id" in Cities_Table_02_df.columns:
    master_df = safe_merge(master_df, Cities_Table_02_df, "city_id")

# Countries
if "country_id" in master_df.columns and "country_id" in Countries_Table_01_df.columns:
    master_df = safe_merge(master_df, Countries_Table_01_df, "country_id")


print("✅ Master DataFrame Created Successfully")
print("Final Shape:", master_df.shape)

master_df.head()
```

```python
# ==========================================
# Creating Columns for additional information
# ==========================================

# 1️⃣ Extract Month & Year from booking_date
if 'booking_date' in master_df.columns:
    master_df['booking_month'] = master_df['booking_date'].dt.month
    master_df['booking_year'] = master_df['booking_date'].dt.year

# 2️⃣ Extract Travel Month
if 'travel_date' in master_df.columns:
    master_df['travel_month'] = master_df['travel_date'].dt.month

# 3️⃣ Calculate Remaining Balance
if 'total_amount' in master_df.columns and 'amount_paid' in master_df.columns:
    master_df['remaining_amount'] = master_df['total_amount'] - master_df['amount_paid']

# 4️⃣ Create Payment Status Category
if 'remaining_amount' in master_df.columns:
    master_df['payment_category'] = master_df['remaining_amount'].apply(
        lambda x: "Fully Paid" if x == 0 else "Pending"
    )

# 5️⃣ Create Booking Season
def season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Summer"
    elif month in [6, 7, 8]:
        return "Monsoon"
    else:
        return "Autumn"

if 'booking_month' in master_df.columns:
    master_df['booking_season'] = master_df['booking_month'].apply(season)

print("✅ New Columns Created Successfully")
master_df.head()
```

```python
# ============================================================
# MATHEMATICAL OPERATIONS (NUMPY + PANDAS)
# ============================================================

# A. Activity Revenue Analysis (Pandas Aggregation)
activity_revenue = activity_bookings.merge(activities, on='activity_id')
rev_summary = activity_revenue.groupby('activity_type')['total_amount'].agg(['sum', 'mean', 'count'])

# B. Population Density Calculation (Vectorized Operation)
# Density = Population / Area
cities['pop_density'] = cities['population'] / cities['area_km2']
top_density = cities[['city_name', 'pop_density']].sort_values(by='pop_density', ascending=False)

# C. Outlier Detection using NumPy (Z-Score)
# Identify bookings that are more than 2 Standard Deviations from the mean
amounts = bookings['total_amount'].to_numpy()
mean_val = np.mean(amounts)
std_val = np.std(amounts)
bookings['z_score'] = (amounts - mean_val) / std_val
outliers = bookings[np.abs(bookings['z_score']) > 2]

# D. Fleet Statistics (NumPy)
fleet_sizes = airlines['fleet_size'].to_numpy()
fleet_avg = np.mean(fleet_sizes)
fleet_max = np.max(fleet_sizes)

# --- DISPLAYING OUTPUTS ---

print("=== 1. REVENUE BY ACTIVITY TYPE ===")
display(rev_summary)

print("\n=== 2. TOP 5 CITIES BY POPULATION DENSITY ===")
display(top_density.head(5))

print("\n=== 3. AIRLINE FLEET STATISTICS ===")
print(f"Average Fleet Size: {fleet_avg:.2f}")
print(f"Largest Fleet: {fleet_max}")

print("\n=== 4. STATISTICAL OUTLIERS IN BOOKINGS ===")
if not outliers.empty:
    display(outliers[['booking_id', 'total_amount', 'z_score']])
else:
    print("No significant outliers detected.")

print("\n=== 5. TOTAL SYSTEM REVENUE ===")
total_rev = activity_bookings['total_amount'].sum() + bookings['total_amount'].sum() + car_rentals['total_cost'].sum()
print(f"Grand Total Revenue: ${total_rev:,.2f}")
```
